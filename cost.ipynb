```python
def apply(path):   
    node = dependencies[path[-1]]
    node.setdefault('use by', set()).add(path[0])
    for depends in node.get("depends on", []): 
        if depends not in path:
            apply([*path, depends])

for name in base_names:
    apply([name])
    
```

```python
def do(name):
    commons = {}
    commons_pkg = {}
    own = 0
    total = 0
    for nname, node in dependencies.items():
        if "use by" not in node or name not in node['use by']:
            continue
        else:
            total += node.get('cost', 0)
            if len(node['use by']) ==1:
                own += node.get('cost', 0)
            else:
                for cname in node['use by']:
                    if cname == name:
                        continue
                    commons.setdefault(cname, 0)
                    commons[cname] += node.get('cost', 0)
                    commons_pkg.setdefault(cname, []).append(nname)
    print("="*30)
    #print(f"{name}: {total}d (own: {own}d, other: {total - own}d)")
    print(f"{name}: {total + global_cost}d")    
    for cname, cost in commons.items():
        pkg = ", ".join(commons_pkg[cname])
        #print()
        #print(f"Shared with {cname}: {cost} ({pkg})")
        print(f"Shared with {cname}: {cost + global_cost}d ({len(commons_pkg[cname])} common elements)")
            

for name in base_names:
    do(name)
    
```
